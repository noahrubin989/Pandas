<ins>Group by</ins>

* The pandas `groupby()` allows us to use what is known as "split-apply-combine"
* First we <b>split</b> our big `df` into smaller independent mini df's
* We then <b>apply</b> aggregation functions of our choice to each of these mini df's, whether this is a mean, median, sum etc.
* After this, we then <b>combine</b> the results and place it into a new (sort of)df with our results
* Generally speaking, this works the same way a groupby works in R's <u>dplyr</u> and in SQL

```python
# Imports and any magic commands
import pandas as pd
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('department_store.csv', parse_dates=['Date'])
df.head()
```

**Exercise: For each city, find total profit generated**

```python
df.groupby('City')['Profit'].sum()
```

**Exercise: For each subcategory of products, what percentage of ratings were less than 7. Plot the results**

```python
# Create new column of booleans
df['rating<7'] = df.RATING < 7

# Groupby this new column. This can be done even though we have a series of booleans! True is 1 and False is 0
(df.groupby('Sub-Category')['rating<7'].mean() *100).plot(kind='bar');
```

**Exercise: For each region and for each segment, find the lowest, highest mean and median product rating**

```python
# Grouping by multiple series requires a list
df.groupby(['Region', 'Segment']).RATING.agg(['min', 'max', 'mean', 'median'])
```

To do this, I'll just see for each month what the total profit and mean profit was
* Use a custom function to create a new variable, Target to see which months returned over 20,000 in profit

```python
def target_hit(values):
    return False if values.sum() < 20_000 else True
```

```python
df.groupby(df.Date.dt.month)['Profit'].agg(Sum=('sum'), Mean=('mean'), Target=target_hit)
```
