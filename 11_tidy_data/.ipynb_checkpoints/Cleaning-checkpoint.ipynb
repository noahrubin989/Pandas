#### Example 1

```python
df = pd.read_csv('../Datasets/peww.csv')  # Allows us to go one directory up.
df.head()
```

- Clearly these columns (apart from religion, which is fine) are values for a variable called income. It also looks like there is a count. 
- We need to create columns called religion, income and count
- df.melt(). Same as the reshape2 package from R

```python
# id_vars is what you don't want to change!
# If you specify id vars, then value vars by default goes with the remaining columns
# Same with value vars. If you select value vars, the id vars will be what is remaining

df.melt(id_vars='religion') 

# Now if we were to name the converted columns melt() also allows this without any functions
df.melt(id_vars='religion',
        var_name='income',
        value_name='count') 
```

#### Example 2

```python
data = pd.read_csv('../Datasets/billboarddata.csv')
data.head()
```

```python
data.melt(id_vars=data.columns[:5],
          value_name='rank',
          var_name='week')
```
